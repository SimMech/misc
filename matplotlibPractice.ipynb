### <font color=red> Histogram </font>

```python
# Fixing random state for reproducibility
np.random.seed(10000000)

x = 100 + 15 * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=True, color='b', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

### <font color=red> Two curves on the same plot </font>

```python
t=np.arange(0,20,1)
```

```python
v1=np.arange(50,70,1)
v1=v1+v1*100*np.random.rand(20)
```

```python
v2=np.arange(30,50,1)
v2=v2+v2*100*np.random.rand(20)
```

```python
plt.plot(t,v1,t,v2)
plt.plot(t,v1,'v',t,v2,'^k')
```
